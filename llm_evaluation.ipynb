```python
import sys
import os
import pandas as pd
import getpass
import torch

!git clone https://github.com/bert-tweeteval

# Add src to path
sys.path.append(os.path.abspath('src'))

from download import download_and_split_dataset
from llm_eval import LLMEvaluator, PROMPT_1_MINIMAL, PROMPT_2_STRUCTURED, LABELS

# Get OpenAI API Key if not set in environment
openai_key = os.getenv("OPENAI_API_KEY")
if not openai_key:
    openai_key = getpass.getpass("Enter your OpenAI API Key: ")
```

```python
# Load dataset
train_df, val_df, test_df = download_and_split_dataset()
print(f"Test set size: {len(test_df)}")

# For evaluation, we use a subset of 100 samples as requested for the benchmark
test_subset = test_df.sample(100, random_state=42)
print(f"Evaluation subset size: {len(test_subset)}")
```

## 1. OpenAI GPT-4o-mini Evaluation

We use the `gpt-4o-mini` model with temperature 0 for deterministic results.

```python
evaluator_openai = LLMEvaluator(openai_api_key=openai_key)

print("Running GPT-4o-mini with Minimal Prompt...")
res_gpt_p1 = evaluator_openai.evaluate(test_subset, "openai", PROMPT_1_MINIMAL, num_samples=None)

print("Running GPT-4o-mini with Structured Prompt...")
res_gpt_p2 = evaluator_openai.evaluate(test_subset, "openai", PROMPT_2_STRUCTURED, num_samples=None)
```

## 2. Qwen3.5-0.8B Evaluation

We run the open-source Qwen model locally.

```python
evaluator_qwen = LLMEvaluator(hf_model_name="Qwen/Qwen3.5-0.8B")
evaluator_qwen.load_hf_model()

print("Running Qwen with Minimal Prompt...")
res_qwen_p1 = evaluator_qwen.evaluate(test_subset, "hf", PROMPT_1_MINIMAL, num_samples=None)

print("Running Qwen with Structured Prompt...")
res_qwen_p2 = evaluator_qwen.evaluate(test_subset, "hf", PROMPT_2_STRUCTURED, num_samples=None)
```

## 3. Results Comparison

We compare Accuracy, Macro-F1, and Inference Time (per 100 samples).

```python
results = {
    "GPT-4o-mini (Minimal)": res_gpt_p1,
    "GPT-4o-mini (Structured)": res_gpt_p2,
    "Qwen-0.8B (Minimal)": res_qwen_p1,
    "Qwen-0.8B (Structured)": res_qwen_p2
}

comparison_df = pd.DataFrame({
    k: {m: v[m] for m in ["Accuracy", "Macro-F1", "Time_per_100"]}
    for k, v in results.items()
}).transpose()

comparison_df.style.highlight_max(subset=["Accuracy", "Macro-F1"], color='lightgreen')
```

## 4. Analysis and Stability

### Prompt Stability Analysis

**Q: Which prompt is more stable and why?**

**Answer:** 
**Structured prompts** significantly out-perform minimal prompts in terms of stability, particularly for smaller open-source models like Qwen-0.8B. Here is why:
1. **Constraint Enforcement**: The structured prompt explicitly instructs the model to `Output ONLY the label name`. This prevents the common issue of models providing reasoning or conversational filler (e.g., "The emotion is joy because...") which can break automated parsing.
2. **Semantic Clarity**: By providing definitions for each label (e.g., `optimism: The tweet expresses hopefulness...`), we anchor the model's understanding of potentially ambiguous terms, leading to more consistent classifications.
3. **Format Consistency**: Structured prompts often use headers like `Label:`, which naturally guides the model's completion behavior toward a single-token response, reducing variance in output length and content.

For larger models like `gpt-4o-mini`, the stability difference is less pronounced as they are highly instruction-tuned, but structured prompts still offer a slight edge in ensuring production-ready parseable outputs.
